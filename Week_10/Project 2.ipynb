```python
import math

class GeometryCalculator:
    def __init__(self):
        pass

    def calculate_trapezium_area(self, base1, base2, height):
        return (height / 2) * (base1 + base2)

    def calculate_rhombus_area(self, diagonal1, diagonal2):
        return 0.5 * diagonal1 * diagonal2

    def calculate_parallelogram_area(self, base, altitude):
        return base * altitude

    def calculate_cube_area(self, side_length):
        return 6 * (side_length ** 2)

    def calculate_cylinder_volume(self, radius, height):
        return math.pi * (radius ** 2) * height

calculator = GeometryCalculator()

print("Welcome to the Geometry Calculator!")

while True:
    print("\nPlease select an equation to calculate:")
    print("1. Area of Trapezium")
    print("2. Area of Rhombus")
    print("3. Area of Parallelogram")
    print("4. Area of Cube")
    print("5. Volume of Cylinder")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == '1':
        base1 = float(input("Enter the length of base 1: "))
        base2 = float(input("Enter the length of base 2: "))
        height = float(input("Enter the height: "))
        area = calculator.calculate_trapezium_area(base1, base2, height)
        print("The area of the trapezium is:", area)
    elif choice == '2':
        diagonal1 = float(input("Enter the length of diagonal 1: "))
        diagonal2 = float(input("Enter the length of diagonal 2: "))
        area = calculator.calculate_rhombus_area(diagonal1, diagonal2)
        print("The area of the rhombus is:", area)
    elif choice == '3':
        base = float(input("Enter the base length: "))
        altitude = float(input("Enter the altitude: "))
        area = calculator.calculate_parallelogram_area(base, altitude)
        print("The area of the parallelogram is:", area)
    elif choice == '4':
        side_length = float(input("Enter the side length: "))
        area = calculator.calculate_cube_area(side_length)
        print("The area of the cube is:", area)
    elif choice == '5':
        radius = float(input("Enter the radius: "))
        height = float(input("Enter the height: "))
        volume = calculator.calculate_cylinder_volume(radius, height)
        print("The volume of the cylinder is:", volume)
    elif choice == '6':
        print("Thank you for using the Geometry Calculator. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a valid choice (1-6).")
```
