```python
from abc import ABC, abstractmethod
import pandas as pd

class External_Vendors(ABC):
    
    @abstractmethod
    def menu(self):
        pass
    
    
class coop_caf(External_Vendors):
    def menu(self):
        print("This is cooperative Menu")
        coop_menu = {
            'Main Meal':['Jollof Rice and Stew','White Rice and Stew','Fried Rice','Salad','Paltain'],
            'Price':['200','200','200','100','100']}
        df = pd.DataFrame(coop_menu)
        df.to_csv('Cooperative_Menu.csv')
        print(df)
        
        
        
class faith_caf(External_Vendors):
    def menu(self):
        print("This Faith Menu")
        faith_menu = {
            'Main Meal':['Fired','White Rice and Stew','Jollof Rice','Beans','Chicken'],
            'Price':['400','400','400','200','1000']}
        df = pd.DataFrame(faith_menu)
        df.to_csv('Faith_Menu.csv')
        print(df)
                

        
class stud_cen_caf(External_Vendors):
    def menu(self):
        print("This is Student Center Menu")
        studcen_menu ={
            'Main Meal':['Chicken Fried','Pomo','Spaghetti','Amala/Ewedu','Semo with Eforio Soup'],
            'Price':['800','300','500','500','500']}
        df = pd.DataFrame(studcen_menu)
        df.to_csv('Student_Center.csv')
        print(df)
       
        
        
        
        
cafteria = input("Which cafeterias Menu would you like to see?: ")

if cafteria == "Cooperative":
    caf_coop = coop_caf()
    caf_coop.menu()
elif cafteria == "Faith":
    caf_faith = faith_caf()
    caf_faith.menu()
elif cafteria == "Student Center":
    caf_stud = stud_cen_caf()
    caf_stud.menu()
else:
    print("Try Again, We do not have that menu")




        
        
        
```
